# Mathematic Operation Using NumPy

```python
import numpy as np
```

```python
arr1= np.arange(1,10).reshape(3,3)
arr2= np.arange(1,10).reshape(3,3)
print(arr1)
print(arr2)
```

addition of two arrays

```python
arr1+arr2
```

substraction of two arrays

```python
arr1-arr2
```

multiplication of two arrays.

```python
arr1*arr2
```

# add()

we can also use add() function for addition of two matrixs.

```python
np.add(arr1,arr2)
```

# subtract() 

we can also use substract function to subtact two matrix.

```python
np.subtract(arr1,arr2)
```

# divide() 

we can also use divide funvtion to divide two matrix.

```python
np.divide(arr1,arr2)
```

# multiply()

we can also use multi function to multiply two matrix.

```python
np.multiply(arr1,arr2)
```

# matrix operation

by using @ we can get product of two matrix (row to column).

```python
arr1@arr2
```

```python
arr1.dot(arr2)
```
