# Python NumPy Array Concatination and Split

```python
import numpy as np
```

```python
arr1=np.arange(1,17).reshape(4,4)
print(arr1)
```

```python
arr2=np.arange(17,33).reshape(4,4)
print(arr2)
```

concatinate two arrays:

```python
list1=[2,3,4,5]
list2=[4,5,6,7]
```

```python
list1+list2
```

```python
arr1+arr2
```

as you can see by using + operator we have addition of two arrays!

```python
np.concatenate((arr1,arr2))
```

```python
print(arr1)
print(arr2)
```

```python
np.concatenate((arr1,arr2), axis=1)
```

```python
np.concatenate((arr1,arr2), axis=0)
```

```python
np.vstack((arr1,arr2))
```

```python
np.hstack((arr1,arr2))
```

we can also concatenate multiple array

```python
np.hstack((arr1,arr2,arr1))
```

```python
np.vstack((arr1,arr2,arr1))
```
